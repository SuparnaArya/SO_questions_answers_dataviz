```python
import pandas as pd
import numpy as np
```

```python
#load metrics for 2 tags csv into dataframes
python_df=pd.read_csv("python_metrics.csv")
#remove entries where timestamps are wrong so difference between question and answering time is negative
python_df=python_df[python_df['first_response_time(minutes)']>=0]
#summarize the metrics data
python_df.describe()

javascript_df=pd.read_csv("javascript_metrics.csv")
javascript_df=javascript_df[javascript_df['first_response_time(minutes)']>=0]
javascript_df.describe()
```

```python
#take means of metrics for 2 tags
python_avg_df=pd.DataFrame(python_df[['view_count','score','answer_count','comment_count','first_response_time(minutes)']].mean().to_dict(),index=[python_df.index.values[-1]])
python_avg_df=python_avg_df.rename(columns=lambda c: "avg_"+str(c)).round(decimals=2)
python_avg_df
javascript_avg_df=pd.DataFrame(javascript_df[['view_count','score','answer_count','comment_count','first_response_time(minutes)']].mean().to_dict(),index=[javascript_df.index.values[-1]])
javascript_avg_df=javascript_avg_df.rename(columns=lambda c: "avg_"+str(c)).round(decimals=2)
javascript_avg_df
#visualize differences between the means in metrics 
import matplotlib.pyplot as plt
fig,ax = plt.subplots(1,5,figsize=(30,14))
x1 = [i-0.1 for i in range(len(python_avg_df))]
x2 = [i+0.1 for i in range(len(javascript_avg_df))]
for i,col in enumerate(python_avg_df.columns):
    ax[i].bar(x=x1,height=python_avg_df[col],width=0.2,label="python")
    for p in ax[i].patches:
        ax[i].annotate(s=np.round(p.get_height(), decimals=2),
                    xy=(p.get_x()+p.get_width()/2., p.get_height()),
                    ha='center',
                    va='center',
                    xytext=(0, 10),
                    textcoords='offset points')
    ax[i].bar(x=x2,height=javascript_avg_df[col],width=0.2,label="javascript")
    for p in ax[i].patches:
        ax[i].annotate(s=np.round(p.get_height(), decimals=2),
                    xy=(p.get_x()+p.get_width()/2., p.get_height()),
                    ha='center',
                    va='center',
                    xytext=(0, 10),
                    textcoords='offset points')
    ax[i].set_title(col)
    ax[i].legend()
```

```python
#take minimum values of metrics for 2 tags
python_min_df=pd.DataFrame(python_df[['view_count','score','answer_count','comment_count','first_response_time(minutes)']].min().to_dict(),index=[python_df.index.values[-1]])
python_min_df=python_min_df.rename(columns=lambda c: "min_"+str(c)).round(decimals=2)
javascript_min_df=pd.DataFrame(javascript_df[['view_count','score','answer_count','comment_count','first_response_time(minutes)']].min().to_dict(),index=[javascript_df.index.values[-1]])
javascript_min_df=javascript_min_df.rename(columns=lambda c: "min_"+str(c)).round(decimals=2)
javascript_min_df
#visualize differences between the minimums for metrics 
fig,ax = plt.subplots(1,5,figsize=(30,14))
x1 = [i-0.1 for i in range(len(python_min_df))]
x2 = [i+0.1 for i in range(len(javascript_min_df))]
for i,col in enumerate(python_min_df.columns):
    ax[i].bar(x=x1,height=python_min_df[col],width=0.2,label="python")
    for p in ax[i].patches:
        ax[i].annotate(s=np.round(p.get_height(), decimals=2),
                    xy=(p.get_x()+p.get_width()/2., p.get_height()),
                    ha='center',
                    va='center',
                    xytext=(0, 10),
                    textcoords='offset points')
    ax[i].bar(x=x2,height=javascript_min_df[col],width=0.2,label="javascript")
    for p in ax[i].patches:
        ax[i].annotate(s=np.round(p.get_height(), decimals=2),
                    xy=(p.get_x()+p.get_width()/2., p.get_height()),
                    ha='center',
                    va='center',
                    xytext=(0, 10),
                    textcoords='offset points')
    ax[i].set_title(col)
    ax[i].legend()
```

```python
#take maximum values of metrics for 2 tags
python_max_df=pd.DataFrame(python_df[['view_count','score','answer_count','comment_count','first_response_time(minutes)']].max().to_dict(),index=[python_df.index.values[-1]])
python_max_df=python_max_df.rename(columns=lambda c: "max_"+str(c)).round(decimals=2)
javascript_max_df=pd.DataFrame(javascript_df[['view_count','score','answer_count','comment_count','first_response_time(minutes)']].max().to_dict(),index=[javascript_df.index.values[-1]])
javascript_max_df=javascript_max_df.rename(columns=lambda c: "max_"+str(c)).round(decimals=2)
#visualize differences between the maximums for metrics 
fig,ax = plt.subplots(1,5,figsize=(30,14))
x1 = [i-0.1 for i in range(len(python_max_df))]
x2 = [i+0.1 for i in range(len(javascript_max_df))]
for i,col in enumerate(python_max_df.columns):
    ax[i].bar(x=x1,height=python_max_df[col],width=0.2,label="python")
    for p in ax[i].patches:
        ax[i].annotate(s=np.round(p.get_height(), decimals=2),
                    xy=(p.get_x()+p.get_width()/2., p.get_height()),
                    ha='center',
                    va='center',
                    xytext=(0, 10),
                    textcoords='offset points')
    ax[i].bar(x=x2,height=javascript_max_df[col],width=0.2,label="javascript")
    for p in ax[i].patches:
        ax[i].annotate(s=np.round(p.get_height(), decimals=2),
                    xy=(p.get_x()+p.get_width()/2., p.get_height()),
                    ha='center',
                    va='center',
                    xytext=(0, 10),
                    textcoords='offset points')
    ax[i].set_title(col)
    ax[i].legend()
```

```python
#get total view counts,answer counts and comment counts for the 2 tags
python_total_df=pd.DataFrame(python_df[['view_count','answer_count','comment_count',]].sum().to_dict(),index=[python_df.index.values[-1]])
python_total_df=python_total_df.rename(columns=lambda c: "total_"+str(c)).round(decimals=2)
javascript_total_df=pd.DataFrame(javascript_df[['view_count','answer_count','comment_count',]].sum().to_dict(),index=[python_df.index.values[-1]])
javascript_total_df=javascript_total_df.rename(columns=lambda c: "total_"+str(c)).round(decimals=2)
#visualize differences between the three metrics 
fig,ax = plt.subplots(1,3,figsize=(30,14))
x1 = [i-0.1 for i in range(len(python_total_df))]
x2 = [i+0.1 for i in range(len(javascript_total_df))]
for i,col in enumerate(python_total_df.columns):
    ax[i].bar(x=x1,height=python_total_df[col],width=0.2,label="python")
    for p in ax[i].patches:
        ax[i].annotate(s=np.round(p.get_height(), decimals=2),
                    xy=(p.get_x()+p.get_width()/2., p.get_height()),
                    ha='center',
                    va='center',
                    xytext=(0, 10),
                    textcoords='offset points')
    ax[i].bar(x=x2,height=javascript_total_df[col],width=0.2,label="javascript")
    for p in ax[i].patches:
        ax[i].annotate(s=np.round(p.get_height(), decimals=2),
                    xy=(p.get_x()+p.get_width()/2., p.get_height()),
                    ha='center',
                    va='center',
                    xytext=(0, 10),
                    textcoords='offset points')
    ax[i].set_title(col)
    ax[i].legend()
```

```python
#load additional csvs for additional analysis for two tags
python_next=pd.read_csv("python_imm.csv")
javascript_next=pd.read_csv("javascript_imm.csv")
python_next['creation_date']=pd.to_datetime(python_next['creation_date'])
python_next['answer_creation_date']=pd.to_datetime(python_next['answer_creation_date'])
javascript_next['creation_date']=pd.to_datetime(javascript_next['creation_date'])
javascript_next['answer_creation_date']=pd.to_datetime(javascript_next['answer_creation_date'])
```

```python
#get total questions,total answers,total unique askers, unique answerers for 2 tags
python_qna=[python_next['id'].nunique(),len(python_next),python_next['asker_user_id'].nunique(),python_next['answerer_user_id'].nunique()]
javascript_qna=[javascript_next['id'].nunique(),len(javascript_next),javascript_next['asker_user_id'].nunique(),javascript_next['answerer_user_id'].nunique()]
#visualize the differences
fig,ax = plt.subplots(1,4,figsize=(20,10))
for i in range(4):
    ax[i].bar(x=-0.2,height=python_qna[i],width=0.2,label="python")
    for p in ax[i].patches:
        ax[i].annotate(s=np.round(p.get_height(), decimals=2),
                    xy=(p.get_x()+p.get_width()/2., p.get_height()),
                    ha='center',
                    va='center',
                    xytext=(0, 10),
                    textcoords='offset points')
    ax[i].bar(x=+0.0,height=javascript_qna[i],width=0.2,label="javascript")
    for p in ax[i].patches:
        ax[i].annotate(s=np.round(p.get_height(), decimals=2),
                    xy=(p.get_x()+p.get_width()/2., p.get_height()),
                    ha='center',
                    va='center',
                    xytext=(0, 10),
                    textcoords='offset points')
ax[0].title.set_text('Total questions answered')
ax[1].title.set_text('Total answers')
ax[2].title.set_text('Total unique questions askers')
ax[3].title.set_text('Total unique answerers')
```

```python
#no. of users who asked and answered for both tags
javascript_q_users=javascript_next[['asker_user_id']]
python_q_users=python_next[['asker_user_id']]
python_a_users=python_next[['answerer_user_id']]
javascript_a_users=javascript_next[['answerer_user_id']]
both_q_users=python_q_users.merge(javascript_q_users,how='inner')
both_q_users['asker_user_id'].nunique()
both_a_users=python_a_users.merge(javascript_a_users,how='inner')
both_a_users['answerer_user_id'].nunique()
fig,ax = plt.subplots(1,2,figsize=(10,6))
ax[0].bar(x=-0.2,height=both_q_users['asker_user_id'].nunique(),width=0.2)
ax[0].set_title('Users who asked questions on both tags')
ax[1].bar(x=-0.2,height=both_a_users['answerer_user_id'].nunique(),width=0.2)
ax[1].set_title('Users who asnswered questions on both tags')

for i in range(2):
    for p in ax[i].patches:
            ax[i].annotate(s=np.round(p.get_height(), decimals=2),
                    xy=(p.get_x()+p.get_width()/2., p.get_height()),
                    ha='center',
                    va='center',
                    xytext=(0, 10),
                    textcoords='offset points')
```

```python
#view trends of questions over time for both tags
python_q=python_next.drop_duplicates(subset='id',keep='first')
javascript_q=javascript_next.drop_duplicates(subset='id',keep='first')
python_qbytime=pd.DataFrame({'python_questions' :python_q['id'].groupby([python_q['creation_date'].dt.year,python_next['creation_date'].dt.month]).size()})
javascript_qbytime=pd.DataFrame({'javascript_questions' :javascript_q.groupby([javascript_q['creation_date'].dt.year,javascript_next['creation_date'].dt.month]).size()})
ax=python_qbytime.plot(kind='line')
javascript_qbytime.plot(ax=ax,kind='line')
ax.set_xlabel('Questioning time')
ax.set_title('Questions over time')
```

```python
#view trends of answers over time for both tags
python_abytime=pd.DataFrame({'python_answers' :python_next['id'].groupby([python_next['answer_creation_date'].dt.year,python_next['answer_creation_date'].dt.month]).size()})
javascript_abytime=pd.DataFrame({'javascript_answers' :javascript_next.groupby([pd.to_datetime(javascript_next['answer_creation_date']).dt.year,pd.to_datetime(javascript_next['answer_creation_date']).dt.month]).size()})
ax=python_abytime.plot(kind='line',label='python answers')
javascript_abytime.plot(ax=ax,kind='line',label='javascript answers')
ax.set_xlabel('Answering time')
ax.set_title('Answers over time')
```
