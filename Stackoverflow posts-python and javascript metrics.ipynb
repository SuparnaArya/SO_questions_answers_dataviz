```python
import pandas as pd
import numpy as np
```

```python
#read data from csv for python tag and keep in a dataframe and explore the columns
python_df=pd.read_csv("python-jan'21-jun'21.csv")
python_df
python_df.columns
```

```python
#rename columns to relevant names
python_df=python_df.rename(columns={'id_1':'answer_id','body_1':'answer_body','owner_user_id':'asker_user_id','owner_user_id_1':'answerer_user_id','creation_date_1':'answer_creation_date',
                          'score_1':'answer_score','favorite_count':'question_favorite_count','favorite_count_1':'answer_favorite_count','comment_count_1':'answer_comment_count','parent_id_1': 'question_id'})
python_df=python_df[['id', 'title', 'body', 'accepted_answer_id', 'answer_count','comment_count','creation_date',
       'question_favorite_count', 'score','asker_user_id','answer_favorite_count','answerer_user_id','tags','view_count','answer_id','answer_creation_date','answer_score','answer_comment_count']]
python_df['creation_date']=pd.to_datetime(python_df['creation_date'])
python_df['answer_creation_date']=pd.to_datetime(python_df['answer_creation_date'])
#remove the answers given after June,2021
python_df=python_df[python_df['answer_creation_date'].dt.month<7]
#calculate response time for each question in minutes
python_df['response_time(minutes)']=(python_df['answer_creation_date'] - python_df['creation_date'])/np.timedelta64(1,'m')
#sorting by response time from lower to higher
python_df=python_df.sort_values(by=['id','response_time(minutes)'])
python_df
```

```python
#generating extra csv for additional analysis
python_imm_df=python_df[['id','creation_date',
       'question_favorite_count', 'score','asker_user_id','answer_favorite_count','answerer_user_id','answer_id','answer_creation_date','answer_score','answer_comment_count']]
python_imm_df.to_csv("python_imm.csv",index=False)
```

```python
#keeping the answers with the earliest response time only
python_df=python_df.drop_duplicates(subset='id',keep='first')
python_df=python_df.rename(columns={'response_time(minutes)':'first_response_time(minutes)'})
#taking views,score,answer count,comments count and earliest response time for a question as metrics to compare with another tag
python_metrics=python_df[['id','view_count','score','answer_count','comment_count','first_response_time(minutes)']]
python_metrics
```

```python
python_metrics.to_csv("python_metrics.csv",index=False)
```

```python
##generating metrics csv for the javascript tag in the same way as python
javascript_df=pd.read_csv("javascript-jan'21-jun'21.csv")
javascript_df=javascript_df.rename(columns={'id_1':'answer_id','body_1':'answer_body','owner_user_id':'asker_user_id','owner_user_id_1':'answerer_user_id','creation_date_1':'answer_creation_date',
                          'score_1':'answer_score','favorite_count':'question_favorite_count','favorite_count_1':'answer_favorite_count','comment_count_1':'answer_comment_count','parent_id_1': 'question_id'})
javascript_df=javascript_df[(pd.to_datetime(javascript_df['answer_creation_date'])).dt.month<7]
javascript_df['response_time(minutes)']=(pd.to_datetime(javascript_df['answer_creation_date'])- pd.to_datetime(javascript_df['creation_date']))/np.timedelta64(1,'m')
javascript_df=javascript_df[(pd.to_datetime(javascript_df['answer_creation_date'])).dt.month<7]
javascript_df=javascript_df.sort_values(by=['id','response_time(minutes)'])
```

```python
javascript_imm_df=javascript_df[['id','creation_date',
       'question_favorite_count', 'score','asker_user_id','answer_favorite_count','answerer_user_id','answer_id','answer_creation_date','answer_score','answer_comment_count']]
javascript_imm_df.to_csv("javascript_imm.csv",index=False)
```

```python
javascript_df=javascript_df.drop_duplicates(subset='id',keep='first')
javascript_df=javascript_df.rename(columns={'response_time(minutes)':'first_response_time(minutes)'})
javascript_metrics=javascript_df[['id','view_count','score','answer_count','comment_count','first_response_time(minutes)']]
javascript_metrics
```

```python
javascript_metrics.to_csv("javascript_metrics.csv",index=False)
```
